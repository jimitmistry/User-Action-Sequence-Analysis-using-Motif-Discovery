# Promoters vs Dropouts

## Promoted 1 vs Dropout 1

### Promoted 1 patterns

```python
df1_new = get_df_set(sax_mtf_p1)
df1_new.loc[df1_new['sum_counts']>np.quantile(df1_new['sum_counts'], 0.9)].head(10)
```

### Dropout 1 patterns

```python
df1_d = get_df_set(sax_mtf_d1)
df1_d.loc[df1_d['sum_counts']>np.quantile(df1_d['sum_counts'], 0.9)].head(10)
```

### Gap analysis: Activities of Promoters not done by Dropouts 

```python
df1_new.loc[df1_new['set1'].isin(set(list(df1_new['set1'].values)) - set(list(df1_d['set1'].values)))].head(10)
```

### Gap Analysis: Activities of Dropouts not done by Promoters

```python
df1_d.loc[df1_d['set1'].isin(set(list(df1_d['set1'].values)) - set(list(df1_new['set1'].values)))].head(10)
```
